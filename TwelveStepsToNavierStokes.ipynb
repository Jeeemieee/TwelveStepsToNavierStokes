```python
#1.OneDimensionalLinearConvection
import numpy as np
from matplotlib import pyplot

file = open("1_LinearConvectionData.txt","r")
data = file.readlines()
x = np.zeros(len(data))
u = np.zeros(len(data))

for i in range(len(data)):
    datasplit = data[i].split(",")
    x[i] = datasplit[0]
    u[i] = datasplit[1]

pyplot.plot(x, u);
```

```python
#5.TwoDimensionalLinearConvection
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot, cm

file = open("5_LinearConvectionData.txt","r")
data = file.readlines()
x = [float(x) for x in data[0].split(",")]
y = [float(x) for x in data[1].split(",")]
u = np.zeros((len(y),len(x)))
for i in range(len(y)):
    u[i] = data[i+2].split(",")   

fig = pyplot.figure(figsize=(10, 10), dpi=100)
ax = fig.gca(projection='3d')                      
X, Y = np.meshgrid(x, y)   
surf = ax.plot_surface(X, Y, u[:], cmap=cm.viridis)
```

```python
#6.TwoDimensionalConvection
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot, cm

file = open("6_ConvectionData.txt","r")
data = file.readlines()
x = [float(x) for x in data[0].split(",")]
y = [float(x) for x in data[1].split(",")]
u = np.zeros((len(y),len(x)))
v = np.zeros((len(y),len(x)))
for i in range(len(y)):
    u[i] = data[i+2].split(",")
    v[i] = data[i+2+len(y)].split(",")
    
fig = pyplot.figure(figsize=(10, 10), dpi=100)
ax = fig.gca(projection='3d')                      
X, Y = np.meshgrid(x, y)   
ax.plot_surface(X, Y, u, cmap=cm.viridis, rstride=2, cstride=2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$');
```

```python
#7_TwoDimensionalDiffusion
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot, cm
import matplotlib.animation as animation

file = open("7_DiffusionData.txt","r")
data = file.readlines()
timeSteps = [int(x) for x in data[0].split(",")]
x = [float(x) for x in data[1].split(",")]
y = [float(x) for x in data[2].split(",")]
u = np.zeros((len(y),len(x)))

print(timeSteps)
timeStep = input("Which timesteps presented above would you like to see?")
index = timeSteps.index(int(timeStep))

for i in range(len(y)):
    u[i] = data[i+3+index*len(y)].split(",")

fig = pyplot.figure(figsize=(10, 10), dpi=100)
ax = fig.gca(projection='3d')                      
X, Y = np.meshgrid(x, y)   
ax.plot_surface(X, Y, u, cmap=cm.viridis, rstride=2, cstride=2, antialiased=True)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlim3d(1, 2.5)
```

```python
#11.CavityFlow
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot, cm

file = open("11_CavityFlow.txt","r")
data = file.readlines()
x = [float(x) for x in data[0].split(",")]
y = [float(x) for x in data[1].split(",")]
X, Y = np.meshgrid(x, y)
u = np.zeros((len(y),len(x)))
v = np.zeros((len(y),len(x)))
p = np.zeros((len(y),len(x)))
for i in range(len(y)):
    u[i] = data[i+2].split(",")
    v[i] = data[i+2+len(y)].split(",")
    p[i] = data[i+2+2*len(y)].split(",")
fig = pyplot.figure(figsize=(11, 7), dpi=100)
pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
pyplot.colorbar()
pyplot.contour(X, Y, p, cmap=cm.viridis)
pyplot.streamplot(X, Y, u, v)
pyplot.xlabel('X')
pyplot.ylabel('Y');
```
